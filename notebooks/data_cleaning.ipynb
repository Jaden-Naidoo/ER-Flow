```python
#import req. libraries
import pandas as pd
import numpy as np

#read csv into data frame
triage_df = pd.read_csv('../data/synthetic_realistic_er_triage.csv')
```

```python
#imputing missing values in columns discovered during EDA
numeric_missing_cols = ['respiratory_rate', 'oxygen_saturation', 'temperature', 'pain_score']
for col in numeric_missing_cols:
    triage_df[col] = triage_df[col].astype(float)
    # Add missing indicator
    triage_df[f'{col}_missing'] = triage_df[col].isna().astype(int)
    # Impute per triage level
    def impute_per_group(series):
        # Compute group median
        group_median = series.median()
        # If all values are NaN in this group, fallback to overall median
        if np.isnan(group_median):
            group_median = triage_df[col].median()
        return series.fillna(group_median)
    
    triage_df[col] = triage_df.groupby('triage_level')[col].transform(impute_per_group)
```

```python
triage_df.to_csv("../data/ER_triage_cleaned.csv", index=False)
print("✅ Cleaned synthetic ER triage dataset saved as CSV!")
```
